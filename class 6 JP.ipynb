```python
attack_mean=data["Attack"].mean()

def set_attack(val):
    if val < attack_mean:
        return "Attack Low"
    elif val==attack_mean:
        return "Attack Neutral"
    else:
        return "Attack High"
```

```python
data["Attack High/Low"]= data["Attack"].apply(set_attack)
data
```
